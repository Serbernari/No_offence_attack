```python
def attack_model(model, input_text, num_added_words):
  np.random.seed(42)
  add_words = np.random.choice(positive, num_added_words)
  input_poisoned = input_text
  for w in add_words: 
    input_poisoned = input_poisoned + f' {w}'
  #return input_poisoned
  return Detoxify(model).predict(input_poisoned)
```

```python
def attack_model_toxigen(tokenizer, model, input_text, num_added_words):
  np.random.seed(42)
  add_words = np.random.choice(positive, num_added_words)
  input_poisoned = input_text
  for w in add_words: 
    input_poisoned = input_poisoned + f' {w}'
  tokenizer.add_special_tokens({'pad_token': '[PAD]'})
  print(input_poisoned)
  tokenized = tokenizer(input_poisoned, truncation=True, padding=True, return_tensors='pt')["input_ids"]
  print(len(tokenized))
  output = model(tokenized)
  
  return output.logits.softmax(dim=-1).tolist()[0][1]
```

```python
def attack_model_ru(tokenizer, model, input_text, num_added_words, show_text = True):
  np.random.seed(42)
  add_words = np.random.choice(positive_ru, num_added_words)
  input_poisoned = input_text
  for w in add_words: 
    input_poisoned = input_poisoned + f' {w}'
  #tokenizer.add_special_tokens({'pad_token': '[PAD]'})
  if show_text:
    print(input_poisoned)
  tokenized = tokenizer(input_poisoned, truncation=True, padding=True, return_tensors='pt')["input_ids"]
  output = model(tokenized)
  
  return output.logits.softmax(dim=-1).tolist()[0][1]
```

```python
def perform_experiments(list_of_models, attack_function, attack_text, amount_of_words):
  for model_name in list_of_models:
    try:

      tokenizer = AutoTokenizer.from_pretrained(model_name)
      model = AutoModelForSequenceClassification.from_pretrained(model_name)

      print(model_name)
      #print('\n')

      # for n in range(10):
      #   print(attack_function(tokenizer, model, attack_text, n))
      
      print('\n')

      for n in range(amount_of_words):
        print(attack_function(tokenizer, model, attack_text, n, show_text = False))

      print('\n')

      print('**********************')
      
    except Exception as e: 
      print(e)


```
